### code below to randomly split , concatenate betas and ses and save it

````python
data_base_dir = '../data_files_hj/simulated_data_bioxcelerate/simulated_data_subset/data/'
````

````python
import numpy as np
np.random.seed(42)
````

````python
import pandas as pd
````

````python
import os
all_dirs = os.listdir(data_base_dir)
np.random.shuffle(all_dirs)
````

````python
train, validate, test =np.split(all_dirs, 
                       [int(.6*len(all_dirs)), int(.8*len(all_dirs))])
````

````python
data_dict = {'train' : train, 'validate' : validate, 'test' : test}
````

````python
for kv in data_dict.keys():

    file_nm_lst = list(data_dict[kv])

    concat_dict_beta = {}
    concat_dict_ses = {}

    for pathval in file_nm_lst:

        r_obj1 = ro.r['readRDS'](os.path.join(data_base_dir,pathval))
        beta_obj = r_obj1.rx2('betas')
        se_obj = r_obj1.rx2('ses')
        cv_idx = r_obj1.rx2('cv_idx')
        # print(pathval)
        # print(beta_obj)
        # print(se_obj)
        # print(cv_idx)

        concat_dict_beta[pathval] = list(beta_obj) + list(cv_idx)
        concat_dict_ses[pathval] = list(se_obj) + list(cv_idx)

        # break

    beta_df = pd.DataFrame.from_dict(concat_dict_beta,orient='index')
    beta_df.columns = [f'beta{i}' if i!=834 else 'cv_idx' for i in range(len(beta_df.columns))]

    beta_df['cluster_num'] = beta_df.index.map(lambda v : int(v.split('_')[-1].split('.')[0]))
    beta_df.index.name = 'file_path'

    se_df = pd.DataFrame.from_dict(concat_dict_ses,orient='index')
    se_df.columns = [f'se{i}' if i!=834 else 'cv_idx' for i in range(len(se_df.columns))]

    se_df['cluster_num'] = se_df.index.map(lambda v : int(v.split('_')[-1].split('.')[0]))
    se_df.index.name = 'file_path'

    beta_df.to_parquet(f'../data/processed/concatenated_data_pq/split_0/betas_{kv}.pq')
    se_df.to_parquet(f'../data/processed/concatenated_data_pq/split_0/se_{kv}.pq')

    beta_df.to_csv(f'../data/processed/concatenated_data_csv/split_0/betas_{kv}.csv')
    se_df.to_csv(f'../data/processed/concatenated_data_csv/split_0/se_{kv}.csv')
````
